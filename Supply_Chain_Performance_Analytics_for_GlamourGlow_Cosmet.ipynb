#### **3. Supplier Benchmarking (Defect Rates & Reliability)**  
**Objective:** Compare suppliers by cost, quality, and lead time.  
**Approach:**  
- **Scatter plot** or **table** with:  
  - **X-axis:** Avg. Lead Time.  
  - **Y-axis:** Defect Rate.  
  - **Bubble size:** Unit Cost.  
- Example: **Supplier C** has low cost but high defect rate (high risk).  

**Insight:**  
- Current sourcing prioritizes cost over reliability, increasing rework.  

**Recommendation:**  
- Shift to **high-reliability suppliers** even at slightly higher costs to reduce defects.
